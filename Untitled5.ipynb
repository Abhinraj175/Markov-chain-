```python
# Install whitebox (if not installed)
!pip install whitebox

import whitebox

# Initialize WhiteboxTools
wbt = whitebox.WhiteboxTools()

# Run fill_missing_data on your DEM
wbt.fill_missing_data(
    i="/content/sample_data/Abhi/dem.tif",
    output="/content/sample_data/Abhi/dem_filled.tif",
    filter=11   # smooth window size
)

print("Voids filled successfully → /content/sample_data/Abhi/dem_filled.tif")
```
